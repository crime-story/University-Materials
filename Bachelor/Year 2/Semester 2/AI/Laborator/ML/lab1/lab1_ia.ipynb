```python
import numpy as np

images = np.zeros((9,400,600))

#a 
for i in range(9):
    image = np.load("D:/Desktop/FMI-UNIBUC/ANUL2/SEMESTRUL 2/Inteligenta Artificiala/lab1/images/car_"+str(i)+".npy")
    images[i] = image

#b calculati suma valorilor pixelilor tuturor imaginilor
print(np.sum(images))

#c suma valorilor pixelilor pt fiecare imagine in parte
print(np.sum(images,axis=(1,2)))

#d indexul imaginii cu suma maxima
print(np.argmax(np.sum(images,axis=(1,2))))

#e imaginea medie
imagine_medie = np.mean(images, axis=0)
#pentru afisarea pozei medie
from skimage import io
io.imshow(imagine_medie.astype(np.uint8))
io.show()

#f deviatia standard
dev_standard = np.std(images)
print(dev_standard)
#print(np.std(images))

#g normalizarea imaginilor
print((images-imagine_medie)/dev_standard)
for image in ((images-imagine_medie)/dev_standard):
  io.imshow(image.astype(np.uint8))
  io.show()

#h decupati fiecare imagine, afisand numai liniile cuprinse intre 200 si 300,
#respectiv coloanele cuprinse intre 280 si 400
print(images[:,200:300, 280:400])
for image in images[:, 200:300, 280:400]:
  io.imshow(image.astype(np.uint8))
  io.show()
```
